```python
import pandas as pd
from pathlib import Path    
```

```python
crime_data = pd.read_csv('crime_data.csv', encoding='ISO-8859-1')
```

```python
crime_data.head()
```

```python
crime_data.keys()
```

```python
crime_data =crime_data.dropna()
```

```python
crime_data.head()
```

```python
crime_data=crime_data.rename(columns={'CMPLNT_NUM':"Complaint ID",
                                       'CMPLNT_FR_DT': "Crime Date",
                                       'CMPLNT_FR_TM':"Complaint Time",
                                         'CMPLNT_TO_DT':"Ending Date",
                                        'CMPLNT_TO_TM':"Ending Time", 
                                        'RPT_DT':"Reported Date",
                                        'KY_CD':"Offense Code",
                                        'OFNS_DESC':"Offense Description",
                                        'PD_CD':"Classification Code",
                                        'PD_DESC':"Classification Description",
                                        'CRM_ATPT_CPTD_CD':"Crime Status",
                                        'LAW_CAT_CD':"Offense Level", 
                                        'JURIS_DESC':"Jurisdiction",
                                        'BORO_NM':"Borough",
                                        'ADDR_PCT_CD':"Precinct",
                                        'LOC_OF_OCCUR_DESC':"Location Description", 
                                        'PREM_TYP_DESC':"Premise Description",
                                        'PARKS_NM':"Park Name",
                                        'HADEVELOPT':"Name of NYCHA Housing", 
                                        'X_COORD_CD':"X Coordinate", 
                                        'Y_COORD_CD':"Y Coordinate",
                                        'Latitude':"Latitude",
                                         'Longitude':"Longitude",
                                        })

crime_data.head()
```

```python
crime_data['Crime Date'] = pd.to_datetime(crime_data['Crime Date'], errors='coerce')
```

```python
crime_data['Month'] = crime_data['Crime Date'].dt.month_name()
crime_data['Day of Week'] = crime_data['Crime Date'].dt.day_name()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#I'm going to count crimes by month
crime_by_month = crime_data['Month'].value_counts().sort_index()
plt.figure(figsize=(12,6))
sns.barplot(x=crime_by_month.index, y=crime_by_month.values)
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Month")
plt.show()
```

```python
#I'm going to count crimes by day of the week
crime_by_day = crime_data['Day of Week'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=crime_by_day.index, y=crime_by_day.values)
plt.xlabel("Day of the Week")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Day of the Week")
plt.show()
```

```python
# Top 10 areas with most crimes
crime_by_borough = crime_data['Borough'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=crime_by_borough.index, y=crime_by_borough.values)
plt.xlabel("Borough")
plt.ylabel("Number of Crimes")
plt.title("Top 10 Borough with the Most Crimes")
plt.show()
```

```python
# I'm going to create a pivot table for crime correlation
crime_corr = crime_data.pivot_table(index="Crime Date", columns="Offense Description", aggfunc="size", fill_value=0)

# Plot correlation matrix
plt.figure(figsize=(12, 6))
sns.heatmap(crime_corr.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Between Different Types of Crimes")
plt.show()
```
